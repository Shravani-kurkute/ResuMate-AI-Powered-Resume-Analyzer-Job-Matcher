```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

np.random.seed(42)
print("✅ Setup complete")
```

```python
df['Salary'] = 0.0

for i in range(len(df)):
    if df.loc[i, 'Placement(Y/N)?'] == 'Placed':
        cgpa = df.loc[i, 'Cgpa']
        internship = df.loc[i, 'Internships(Y/N)']
        project = df.loc[i, 'Innovative Project(Y/N)']
        technical = df.loc[i, 'Technical Course(Y/N)']
        training = df.loc[i, 'Training(Y/N)']
        backlog = df.loc[i, 'Backlog in 5th sem']
        communication = df.loc[i, 'Communication level']
        
        cgpa_normalized = cgpa / 10 if cgpa > 10 else cgpa
        base_salary = 5.0
        
        # CGPA bonus (30% weight)
        if cgpa_normalized >= 9.0:
            base_salary += np.random.uniform(4, 8)
        elif cgpa_normalized >= 8.0:
            base_salary += np.random.uniform(2, 6)
        elif cgpa_normalized >= 7.0:
            base_salary += np.random.uniform(1, 4)
        else:
            base_salary += np.random.uniform(0, 2)
        
        # Skills bonus (40% weight) - NOW STRONGER!
        if internship == 'Yes': 
            base_salary += np.random.uniform(2, 5)
        if project == 'Yes': 
            base_salary += np.random.uniform(3, 6)
        if technical == 'Yes': 
            base_salary += np.random.uniform(1, 3)
        if training == 'Yes': 
            base_salary += np.random.uniform(1, 3)
        
        # Communication (20% weight)
        base_salary += communication * 1.0
        
        # Backlog penalty
        if backlog == 'Yes': 
            base_salary -= np.random.uniform(2, 4)
        
        base_salary = max(3.5, base_salary) * np.random.uniform(0.9, 1.1)
        df.loc[i, 'Salary'] = round(base_salary, 1)

print(df[df['Salary'] > 0]['Salary'].describe())
```

```python
def categorize_package(row):
    if row['Salary'] == 0: return 'Not Placed'
    elif row['Salary'] >= 20: return 'Premium'
    elif row['Salary'] >= 15: return 'Standard'
    else: return 'Basic'
df['Package_Category'] = df.apply(categorize_package, axis=1)
print(df['Package_Category'].value_counts())
```
